```python
import pandas as pd
data = pd.read_csv("/Users/blackwingedkite/Desktop/IRTM_project/qapair/百科-001_qa_source.csv")
data2 = pd.read_csv("/Users/blackwingedkite/Desktop/IRTM_project/qapair/百科-002_qa_source.csv")
# 確保 `id` 列存在並將其值加 100
if 'docid' in data2.columns:
    data2['docid'] += 1000000
concat_data = pd.concat([data, data2], axis=0, ignore_index=True)
concat_data
```

```python
docid = concat_data["docid"].tolist()
```

```python
from collections import Counter
# 使用 Counter 計算每個元素出現次數
counter = Counter(docid)

# 根據出現次數由高到低排序
sorted_items = sorted(counter.items(), key=lambda x: x[1], reverse=True)
doc_list = []
# 打印結果
i = 0
for item, count in sorted_items:
    if count>1:
        print(item, count)
    doc_list.append(item)
    i+= 1
    if i == 1000:
        break
print(len(doc_list))
```

```python
concat_source = concat_source[concat_source['id'].isin(doc_list)]
concat_source.shape
```
