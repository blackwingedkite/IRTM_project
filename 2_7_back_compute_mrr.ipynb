```python
from supabase import create_client, Client
import os
import pandas as pd

model_name = "text-embedding-3-small"
openai_api_key = os.environ['OPENAI_API_KEY']
supabase_url = os.environ['SUPABASE_URL']
supabase_api_key = os.environ['SUPABASE_API_KEY'] 
supabase: Client = create_client(supabase_url, supabase_api_key)
```

```python
page_size = 1000
offset = 0
all_data = []
while True:
    response = supabase.table('multidoc_questions').select("id, question, answer, docid, gpt_answer, similar_doc_ids, similar_doc_ids_bm25, similar_doc_ids_tfidf").range(offset, offset + page_size - 1).execute()
    if not response.data:
        break
    all_data.extend(response.data)
    offset += page_size
gpt_answer_result = pd.DataFrame(all_data)
```

```python
page_size = 1000
offset = 0
all_data = []
while True:
    response = supabase.table('multidoc_sources').select("id, docid, title, content").range(offset, offset + page_size - 1).execute()
    if not response.data:
        break
    all_data.extend(response.data)
    offset += page_size
gpt_source_result = pd.DataFrame(all_data)
```

```python
# Similar_doc_ids to document_ids
import time
start = time.time()
gpt_answer_result['similar_origin_doc_ids'] = None
for index, row in gpt_answer_result.iterrows():
    similar_doc_ids = row['similar_doc_ids']
    if similar_doc_ids:
        similar_origin_doc_ids = []
        for chunk_id in similar_doc_ids:
            original_id = gpt_source_result[gpt_source_result['id'] == chunk_id]["docid"].values[0]
            similar_origin_doc_ids.append(gpt_source_result[gpt_source_result['id'] == chunk_id].iloc[0]['docid'])
        gpt_answer_result.at[index, 'similar_origin_doc_ids'] = similar_origin_doc_ids

end = time.time()
print(f"Time taken : {end - start} seconds")
```

```python
# calculate hit rate and mrr
hit_num = 0
mrr_num = 0
def find_index(arr, target):
  try:
      index = arr.index(target, 0)
      return index
  except ValueError:
      return "not_found"
for index, row in gpt_answer_result.iterrows():
    similar_origin_doc_ids = row['similar_origin_doc_ids']
    gold_standard = row['docid']
    print(similar_origin_doc_ids)
    print(gold_standard)
    if gold_standard in similar_origin_doc_ids:
        hit_num += 1
        position = find_index(similar_origin_doc_ids, gold_standard)
        if position == "not_found":
            score = 0
        else:
            score = 1 / (position+1)
        mrr_num += score
hit_rate = hit_num / len(gpt_answer_result)
mrr = mrr_num / len(gpt_answer_result)
print(f"Hit rate: {hit_rate}")
print(f"MRR: {mrr}")
```
